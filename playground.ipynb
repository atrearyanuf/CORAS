```python
# Download NLTK data (for the first-time use)
nltk.download('punkt')

# Initialize the S3 resource
s3 = boto3.resource('s3')

# Access the S3 bucket and object (the .txt file)
obj = s3.Bucket('datacollection1').Object('biden-sotu-2023-planned-official.txt').get()

# Read the file content from the object
file_content = obj['Body'].read().decode('utf-8')

# Tokenize the content into words
tokens = word_tokenize(file_content)

# Print the tokens
print(tokens)
```

```python
from collections import Counter
import string
```

```python
# Tokenize the content into words
tokens = word_tokenize(file_content)

# Remove punctuation and convert to lowercase
cleaned_tokens = [word.lower() for word in tokens if word.isalnum()]

# Create a Counter object to count word frequencies
word_counts = Counter(cleaned_tokens)

# Get the top 5 most common words
top_5_words = word_counts.most_common(5)

# Print the top 5 words and their counts
print("Top 5 most common words:")
for word, count in top_5_words:
    print(f"'{word}': {count} times")
```
